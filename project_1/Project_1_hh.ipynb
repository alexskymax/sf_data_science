1). Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education_type(data):
    """Функция выделяет из строки тип образования соискателя.
    
    Args:
        строку (str): информация об образовании и заведениях где учился соискатель.
    
    Return:
        строку (str): тип образования соискателя.
    """
    # Разбиваем строку по " " на объекты и заносим срез из трёх строк в список
    educ_str = ' '.join(data.split(' ')[:3])
    # Проверяем строку на наличие нужных объектов и возвращаем тип образования
    if 'Высшее' in educ_str:
        return 'высшее'
    elif 'Неоконченное' in educ_str:
        return 'неоконченное высшее'
    elif 'Среднее специальное' in educ_str:
        return 'среднее специальное'
    elif 'Среднее' in educ_str:
        return 'среднее'
    
# Создаём новый признак и применяем нашу функцию к столбцу 'Образование и ВУЗ'
df_hh['Образование'] = df_hh['Образование и ВУЗ'].apply(get_education_type)
# Удаляем признак 'Образование и ВУЗ'
df_hh.drop(columns='Образование и ВУЗ', inplace=True)

# Выводим информацию о количестве уникальных значений в столбце "Образование"
df_hh['Образование'].value_counts()
```

2). Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
def get_sex_age(data):
    """Функция выделяет из строки пол соискателя и возраст.
    
    Args:
        строку (str): информация о возрасте и гендерной принадлежности соискателя.
    
    Return:
        кортеж (tuple): на первом месте строка (str) - пол соискателя в виде "М" или "Ж";
                        на втором месте число (int) - возраст.
    """
    sex = None
    # Создаём переменную и условие, если слово есть в аргументе,
    # присваеваем переменной соответствующее значение
    if 'Мужчина' in data:
        sex = 'М'
    elif 'Женщина' in data:
        sex = 'Ж'
    
    age = None # Создаём переменную age
    data_splitted = data.split(' ') # Разбиваем строку по " "
    year_words=['год', 'года', 'лет']
    for index, item in enumerate (data_splitted): # С помощью функции enumerate проверяем значения в цикле
        if item in year_words:
            age = int(data_splitted[index-1]) # Если значение совпадает с элементом списка year_words, 
                                              # переменной присваиваем предыдущее значение индекса
    
    return sex, age # Возвращаем обе переменные в кортеже

# Создаём признаки "Пол" и "Возраст" и применяем нашу функцию get_sex_age к столбцу "Пол, возраст"
df_hh[['Пол', 'Возраст']] = pd.DataFrame(df_hh['Пол, возраст'] \
    .apply(get_sex_age).tolist(), index=df_hh.index)

# Удаляем признак "Пол, возраст"
df_hh.drop(columns='Пол, возраст', inplace=True)

# Выводим процентное соотношение признака "Пол"
display(round(df_hh['Пол'].value_counts(normalize=True)*100, 2))

# Выводим средний возраст соискателей
display(round(df_hh['Возраст'].mean(), 1))
```

3). Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(arg):
    """Функция выделяет из строки общий стаж соискателя в месяцах.
    
    Args:
        строку (str): информация о стаже и периодах работы в различных компаниях.
    
    Return:
        число (float): общий стаж в месяцах
    """
    if arg is np.nan or arg == 'Не указано': # Проверяем на пропуски и заменяем "Не указано" на None
        return None
    
    month_key_words = ['месяц', 'месяцев', 'месяца']
    year_key_words = ['год', 'лет', 'года'] # Создаём списки с контрольными словами года и месяца
    
    args_splited = arg.split(' ')[:7] # Разбиваем строку по пробелам и делаем срез на первых 6 значениях
    month = 0
    year = 0
    
    for i in range(len(args_splited)): # Запускаем цикл проверки нужных строк в списке
        if args_splited[i] in month_key_words:
            month = args_splited[i-1]
        if args_splited[i] in year_key_words: # если объект в списке добавляем предыдущее значение в переменную
            year = args_splited[i-1]
    return int(year)*12 + int(month) # возвращаем объект в виде количества месяцев

# Создаём признак "Опыт работы (месяц)" и применяем нашу функцию get_experience к столбцу "Опыт работы"
df_hh['Опыт работы (месяц)'] = df_hh['Опыт работы'].apply(get_experience)
# Удаляем признак "Опыт работы"
df_hh.drop(columns='Опыт работы', inplace=True)

# Выводим медианное значение нового признака "Опыт работы (месяц)"
df_hh['Опыт работы (месяц)'].median()
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Инфорация о метро, рядом с которым проживает соискатель нас не интересует.

    * … , готов к переезду , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city_moving_trips(arg):
    """Функция выделяет из строки город проживания, готовность или не готовность
       к переезду и готовность или не готовность к командировкам.
    
    Args:
        строку (str): информация о городе проживания, готовность или не готовность
        к переезду и готовность или не готовность к командировкам..
    
    Return:
        кортеж (tuple): на первом месте строка (str) - город проживания;
                        на втором месте логический (bool) - готовность\не готовность к переезду;
                        на третьем месте логический (bool) - готовность\не готовность к командировкам.
    """
    city = None # Создаём переменную city и список городов миллионников
    million_cities = [
        'Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск',
        'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград'
    ]  
    incoming_arg = arg.split(',')
    city_name = (incoming_arg[0].split(' '))[0] # Разбиваем строку по "," и ещё раз разбиваем первый элемент по " "
    # Создаём условие: если переменная city_name в списке миллионников, Москва или Санкт-Петербург,
    # соответствующее значение присваиваем переменной city
    if city_name in million_cities: city = city_name
    elif city_name == 'Москва': city = 'Москва'
    elif city_name == 'Санкт-Петербург': city = 'Санкт-Петербург'
    else: city = 'другие' # если условие не соблюдается переменной city присваивается значение "другие"
    
    # Создаём переменную moving и условие: если в строке отрицание готовности к переезду
    # переменной присваивается булево значение False, если готовность присваивается True
    moving = None 
    if ('не готов к переезду' in arg) | ('не готова к переезду' in arg):
        moving = False
    elif ('готов к переезду' in arg) | ('готова к переезду' in arg) | ('хочу' in arg):
        moving = True
    
    # Создаём переменную trips и условие: если в строке отрицание готовности к командировкам
    # переменной присваивается булево значение False, если готовность присваивается True
    trips = None
    if ('не готов к командировкам' in arg) | ('не готова к командировкам' in arg):
        trips = False
    elif('готов к командировкам' in arg) | ('готова к командировкам' in arg):
        trips = True
    elif ('готов к редким командировкам' in arg) | ('готова к редким командировкам' in arg):
        trips = True 
    else:
        trips = False
        
    return city, moving, trips # возвращаем кортеж из трёх переменных

# Создаём новые признаки и применяем нашу функцию get_city_moving_trips к столбцу "Город, переезд, командировки"
df_hh[['Город', 'Готовность к переезду', 'Готовность к командировкам']] \
    = pd.DataFrame(df_hh['Город, переезд, командировки'].apply(get_city_moving_trips).tolist(), index=df_hh.index)
# Удаляем признак "Город, переезд, командировки"
df_hh.drop(columns='Город, переезд, командировки', inplace=True)
# Выводим процент соискателей по городам
display(round(df_hh['Город'].value_counts(normalize=True)*100))
# Выводим процент соискателей готовых к командировкам и переезду
display(round(df_hh[(df_hh['Готовность к переезду'] == True) & \
      (df_hh['Готовность к командировкам'] == True)].shape[0]/df_hh.shape[0]*100))
```

```python
def get_schedule(arg):
    """Функция выделяет из строки режимы графика и с помощью условия присваивает соответствующим 
        переменным булево значение.
    
    Args:
        строку (str): информация о режиме графика.
    
    Return:
        кортеж (tuple): из пяти булевых значений.
    """
    flex = None
    full = None
    change = None
    rotate = None
    online = None
    
    my_schedule = ' '.join(arg.split(','))
    
    if 'гибкий график' in my_schedule: flex = True
    else: flex = False    
    if 'полный день' in my_schedule: full = True
    else: full = False   
    if 'сменный график' in my_schedule: change = True
    else: change = False
    if 'вахтовый метод' in my_schedule: rotate = True
    else: rotate = False
    if 'удаленная работа' in my_schedule: online = True
    else: online = False
        
    return flex, full, change, rotate, online

def get_busy(arg):
    """Функция выделяет из строки режимы занятости и с помощью условия присваивает соответствующим 
        переменным булево значение.
    
    Args:
        строку (str): информация о режиме занятости.
    
    Return:
        кортеж (tuple): из пяти булевых значений.
    """
    part = None
    project = None
    full_busy = None
    intern = None
    volunteer = None
    
    my_busy = ' '.join(arg.split(','))
    
    if 'частичная занятость' in my_busy: part = True
    else: part = False    
    if 'проектная работа' in my_busy: project = True
    else: project = False   
    if 'полная занятость' in my_busy: full_busy = True
    else: full_busy = False
    if 'стажировка' in my_busy: intern = True
    else: intern = False
    if 'волонтерство' in my_busy: volunteer = True
    else: volunteer = False
        
    return part, project, full_busy, intern, volunteer

# Создаём новые признаки из переменных режимов графика и занятости, применяем функции к признакам "График" и "Занятость"
df_hh[['гибкий график', 'полный день', 'сменный график', 'вахтовый метод', 'удаленная работа']] \
    = pd.DataFrame(df_hh['График'].apply(get_schedule).tolist(), index=df_hh.index)
df_hh[['частичная занятость', 'проектная работа', 'полная занятость', 'стажировка', 'волонтерство']] \
    = pd.DataFrame(df_hh['Занятость'].apply(get_busy).tolist(), index=df_hh.index)
df_hh.drop(columns=['График', 'Занятость'], inplace=True) # Удаляем ненужные столбцы

# Выводим количество соискателей с нужным фильтром
display(df_hh[(df_hh['волонтерство'] == True) & \
      (df_hh['проектная работа'] == True)].shape[0])
display(df_hh[(df_hh['гибкий график'] == True) & \
      (df_hh['вахтовый метод'] == True)].shape[0])
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".

```python
df_currency = pd.read_csv('data/ExchangeRates.csv', sep=',') # импортруем таблицу с курсами валют
# Приводим столбцы с датами в таблицах к общему виду.
df_currency['date'] = pd.to_datetime(df_currency['date'], dayfirst=True).dt.date
df_hh['Обновление резюме'] = pd.to_datetime(df_hh['Обновление резюме'], dayfirst=True).dt.date

def currency_change(arg):
    """Функция выделяет из строки сумму желаемой зарплаты и валюту.
    
    Args:
        строку (str): информация о желаемой зарплате в различной валюте.
    
    Return:
        кортеж (tuple): из двух элементов, первый сумма желаемой зарплаты (int),
        второй строка(str) обозначение валюты.
    """
    official_curr = ['USD', 'EUR', 'KZT', 'KGS', 'BYN', 'UAH', 'UZS', 'AZN', 'RUB'] # список официальных обозночений валюты
    
    amount = None
    currency = None
    
    currency_lst = arg.split(' ') # Разбиваем строку по " " и создаём список из значений строки
    amount = int(currency_lst[0]) # присваиваем переменной первое значение списка (сумма)
    
    if currency_lst[-1] in official_curr: # проверяем второй элемент списка на наличие в списке официальных валют
        currency = currency_lst[-1]  # присваиваем официальное название валюты переменной
    elif currency_lst[-1] == 'руб.': currency = 'RUB' # меняем произвольное название валюты на официальное 
    elif currency_lst[-1] == 'бел.руб.': currency = 'BYN'
    elif currency_lst[-1] == 'грн.': currency = 'UAH'
    elif currency_lst[-1] == 'сум': currency = 'UZS'
    
    return amount, currency # возвращаем кортеж с переменными(сумма, валюта)

# Создаём новые признаки "Сумма" и "Валюта", применяем функцию к столбцу "ЗП".
df_hh[['Сумма', 'Валюта']] = pd.DataFrame(df_hh['ЗП'].apply(currency_change).tolist(), index=df_hh.index)
# Присоединяем к таблице с резюме, таблицу с курсами валют по признаку "Валюта"("currency") и " Обновление резюме"("date")
df_hh = df_hh.merge(df_currency, 
         left_on=['Обновление резюме','Валюта'], right_on=['date', 'currency'], how='left')
# Заполняем все пропуски (значение курса рубля к рублю) в столбце с курсом ("close") на еденицу
df_hh['close'] = df_hh['close'].fillna(1)
# Заполняем все пропуски (значение пропорции рубля к рублю) в столбце с пропорцией ("proportion") на еденицу
df_hh['proportion'] = df_hh['proportion'].fillna(1)
# Создаём овый признак где выщитываем сумму ЗП в рублях
df_hh['ЗП (руб)'] = (df_hh['Сумма'] * df_hh['close']) / df_hh['proportion']
# Удаляем все не нужные столбцы
df_hh.drop(columns=['ЗП', 'Сумма', 'Валюта', 'currency',\
                       'per', 'date', 'time', 'close', 'vol', 'proportion'], inplace=True)

# Выводим медиану зарплаты в рублях и округлянм до тысячи
round(df_hh['ЗП (руб)'].median()*0.001)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4)) # Создаём объект диаграммы
log_age = np.log(clean_hh_age['Возраст'] + 1) # Логарифмический масштаб для признака "Возраст"
histplot = sns.histplot(log_age, bins=30, ax=ax) # Создаём гистограмму
histplot.axvline(log_age.mean(), color='k', lw=2) # Наносим на гистограмму линию со средним значением
histplot.axvline(log_age.mean()+ 3 * log_age.std(), color='k', ls='--', lw=2) # границы интервала трёх сигм
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Распределение признака "Возраст"');

def outliers_z_score(data, feature, log_scale=False):
    """Функция находит выбросы с помощью метода z-отклонения.
    
    Args:
        DataFrame: первый элемент это таблица, второй это признак таблицы к которой будет применена функция.
    
    Return:
        кортеж (tuple): из двух элементов, первый таблица с выбросами,
        второй таблица после удаления выбросов.
    """
    if log_scale: # прверяем условие для выбора логарифмического масштаба
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score(clean_hh_age, 'Возраст', log_scale=True)
print(outliers['Возраст'])
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
```
